```python
from Calender_agent import schedule_meeting_for_task
from utils.google_auth import get_google_creds
from googleapiclient.discovery import build

def get_google_doc_content(doc_id):
    creds = get_google_creds(['https://www.googleapis.com/auth/documents.readonly'])
    service = build('docs', 'v1', credentials=creds)

    doc = service.documents().get(documentId=doc_id).execute()
    text = ""
    for element in doc.get("body").get("content"):
        if "paragraph" in element:
            for run in element["paragraph"].get("elements", []):
                text += run.get("textRun", {}).get("content", "")
    return text

def extract_task_lines_from_text(doc_text):
    lines = doc_text.splitlines()
    tasks = [line.strip("-â€¢ ").strip() for line in lines if line.strip().lower().startswith(("bloom:", "action item", "- bloom", "â€¢ bloom", "â€¢"))]
    return tasks

# === Step 1: Paste Google Doc ID ===
doc_id = input("ðŸ“„ Paste the Google Doc ID (from the docs_agent):\n")

# === Step 2: Read doc content
doc_text = get_google_doc_content(doc_id)
print("\nðŸ“„ Google Doc content preview:\n", doc_text)

# === Step 3: Extract tasks (rudimentary rule-based)
task_lines = extract_task_lines_from_text(doc_text)

# === Step 4: Schedule any valid task
print("\nâœ… Detected Task Lines:")
for task in task_lines:
    print(" -", task)
    result = schedule_meeting_for_task(task)
    print(result)
```
